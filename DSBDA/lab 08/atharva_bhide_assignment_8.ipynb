```python
def fun1(value):
    if (value == "male"):
        return 1
    else:
        return 0
```

```python
def fun2(value):
    if (value == 'S'):
        return 0
    elif (value == 'C'):
        return 1
    elif (value == 'Q'):
        return 2
    else:
        return 0
```

```python
df["sex"] = df["sex"].apply(fun1)
```

```python
df["embarked"] = df["embarked"].apply(fun2)
```

```python
df.isna().sum()
```

```python
df = df.drop("deck", axis=1)
```
